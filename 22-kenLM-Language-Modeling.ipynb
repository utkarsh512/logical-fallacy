```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

```python
def strip_array(a):
    idx = 0
    while a[idx] == 0:
        idx += 1 
    return a[idx:]
```

```python
from numpy.core.multiarray import can_cast
ad_c = dict()
none_c = dict()
for cat in categories_selected:
    ad_c[cat] = [0]
    none_c[cat] = [0]
for cat in categories_selected:
    for year in range(min_year, max_year + 1):
        for q in range(1, 5):
            ad_c[cat].append(ah[f'{year}_Q{q}_{cat}'])
            none_c[cat].append(none[f'{year}_Q{q}_{cat}'])
    ad_c[cat] = list(accumulate(ad_c[cat], lambda x, y: y or x))
    none_c[cat] = list(accumulate(none_c[cat], lambda x, y: y or x))
    ad_c[cat] = strip_array(ad_c[cat])
    none_c[cat] = strip_array(none_c[cat])
    ad_c[cat] = moving_average(ad_c[cat], n=4) 
    none_c[cat] = moving_average(none_c[cat], n=4)

quarters = list()
for year in range(2008, 2022):
    for q in range(1, 5):
        val = f'{year}_Q{q}'
        quarters.append(val)
```
