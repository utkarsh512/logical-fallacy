# Benchmarking time required to scrape FB for possible configs

```python
import math

def getPostTime(x):
    return math.ceil(x / 5) * 10

def getCommentTime(x):
    return x 

def getReplyTime(x):
    return math.ceil(x / 8) * 10
```

```python
def convertTime(t):
    unit = "secs"
    f = 12 * 28 * 24 * 60 * 60
    if t > f:
        t /= f
        unit = "years"
    f /= 12
    if t > f:
        t /= f
        unit = "months"
    f /= 28
    if t > f:
        t /= f
        unit = "days"
    f /= 24
    if t > f:
        t /= f
        unit = "hrs"
    f /= 60
    if t > f:
        t /= f
        unit = "mins"
    assert t <= 60
    return t, unit
```

```python
s = []

for postCount in [100, 500, 1000, 2000]:
    for commentCount in [100, 500, 1000, 5000]:
        for replyCount in [0, 10, 50, 100]:
            timeTaken = 0
            timeTaken += getPostTime(postCount)
            timeTaken += postCount * getCommentTime(commentCount)
            timeTaken += postCount * commentCount * getReplyTime(replyCount)
            timeProcessed, unit = convertTime(timeTaken)
            val = f"{int(timeProcessed * 100) / 100} {unit}"
            postCount_ = str(postCount)
            commentCount_= str(commentCount)
            replyCount_= str(replyCount)
            dump = f"{postCount_:12}\t{commentCount_:12}\t{replyCount_:12}\t{val}"
            s.append((timeTaken, dump))
```

```python
s = sorted(s)
```

```python
print(f"{'#Post':12}\t{'#Comments':12}\t{'#Replies':12}\t{'Time':12}")
for x in s:
    print(x[1])
```
