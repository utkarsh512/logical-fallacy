## Training and testing

```python
%%shell
cd /content/gdrive/MyDrive/ganbert/
python -u ganbert.py \
        --num_classes=3 \
        --label_rate=0.02 \
        --do_train=false \
        --do_eval=false \
        --do_predict=true \
        --data_dir=data \
        --vocab_file=cased_L-12_H-768_A-12/vocab.txt \
        --bert_config_file=cased_L-12_H-768_A-12/bert_config.json \
        --init_checkpoint=cased_L-12_H-768_A-12/bert_model.ckpt \
        --max_seq_length=64 \
        --train_batch_size=64 \
        --learning_rate=2e-5 \
        --num_train_epochs=3 \
        --warmup_proportion=0.1 \
        --do_lower_case=false \
        --output_dir=ganbert_output_model
```

```python
%%shell
cd /content/gdrive/MyDrive/ganbert/
python -u ganbert.py \
        --num_classes=3 \
        --label_rate=0.02 \
        --do_train=false \
        --do_eval=false \
        --do_predict=true \
        --data_dir=data \
        --vocab_file=cased_L-12_H-768_A-12/vocab.txt \
        --bert_config_file=cased_L-12_H-768_A-12/bert_config.json \
        --init_checkpoint=ganbert_output_model/model.ckpt-393 \
        --max_seq_length=64 \
        --train_batch_size=64 \
        --learning_rate=2e-5 \
        --num_train_epochs=3 \
        --warmup_proportion=0.1 \
        --do_lower_case=false \
        --output_dir=ganbert_output_model
```
