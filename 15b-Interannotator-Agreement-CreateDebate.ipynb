```python
import numpy as np
import pandas as pd
```

```python
from statsmodels.stats.inter_rater import fleiss_kappa, aggregate_raters
```

```python
cols = ['Do you think this is an ad-hominem comment?']
for i in range(1, 20):
    cols.append(cols[0] + '.' + str(i))
```

```python
addr = '/content/gdrive/MyDrive/DL/CreateDebate/Politics/annotations/batch{}.csv'
```

```python
y = []
pids = []

for i in range(1, 6):
    df = pd.read_csv(addr.format(i))
    pids.append(df['Enter your Prolific ID'].tolist())
    for x in cols:
        lbl = df[x].tolist()
        lbl_numeric = []
        for z in lbl:
            if z == 'Yes':
                lbl_numeric.append(1)
            else:
                lbl_numeric.append(0)
        y.append(lbl_numeric)
```

```python
masks = [0, 1, 4] #3, 2
```

```python
labels = []
```

```python
for i in range(4 * 20, 5 * 20):
    labels_ = []
    for j in range(3):
        labels_.append(y[i][masks[j]])
    labels.append(labels_)
```

```python
labels
```

```python
table, _ = aggregate_raters(labels)
```

```python
v = fleiss_kappa(table)
```
